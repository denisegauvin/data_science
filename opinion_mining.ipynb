```python
import spacy
nlp = spacy.load("en_core_web_sm")
import pandas as pd
```

```python
apps = pd.read_csv('data/googleplaystore.csv')
apps.head()
```

```python
reviews = pd.read_csv('data/googleplaystore_user_reviews.csv')
reviews.head()
```

```python
#clean column headers
apps.columns = apps.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
apps.head()
```

```python
#clean column headers
reviews.columns = reviews.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
reviews.head()
```

```python
dataset = pd.merge(reviews, apps, how='right', on='app')

dataset.head(10)
```

```python
dataset[dataset.translated_review.isna()]
```

```python
dataset = dataset.dropna(subset=['translated_review'])
dataset.head()
```

```python
dataset['category'].nunique()
```

```python
#To pick a category, checked number of reviews by category
dataset.groupby('category')['reviews'].count()
```

```python
dataset.groupby('category').first()
```

```python
#selected health and fitness category to explore
dataset_health_and_fitness = dataset[dataset['category'] == 'HEALTH_AND_FITNESS']
dataset_health_and_fitness.head()
```

```python
#To pick one app For future aspect mining, checked number of reviews by app
dataset_health_and_fitness.groupby('app')['reviews'].count()
```
